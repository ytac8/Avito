```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('../data/unzipped/train.csv')
```

```python
test = pd.read_csv('../data/unzipped/test.csv')
```

```python
dataframe = pd.concat([data, test])
```

```python
def label_encode(series):
    item_dict = {}
    for i, s in series.iteritems():
        item_dict[s] = str(i)
    return item_dict
```

```python
item_dict = label_encode(dataframe.item_id)
```

```python
import joblib
```

```python
joblib.dump(item_dict, '../data/pickle/label_dict.pkl', compress=3)
```
